```python
import requests # hitting a URL
from bs4 import BeautifulSoup # parsing and filtering/extration of information
import pandas as pd
from selenium import webdriver
import os
```

```python
# This starts an instance of the Chrome browser to be controlled by the selenium webdriver
base_url = 'https://www.express.pk/crime/archives/'
# driver = webdriver.Chrome()

crime_news = []
data = []
for i in range(90,96):
    target_url = base_url + '?page=' +  str(i)
    print(target_url)
    response = requests.get(target_url)
    soup = BeautifulSoup(response.content, 'html.parser')
    crime_news = [j.select_one("div a:nth-of-type(2)").text for j in soup.find_all('div', {'class':'story cat-0 group-0 position-0 cstoreyitem couplet clearfix'})]
    data.extend(crime_news)
print(data)
```

```python
df = pd.DataFrame(data)

# if file does not exist write header 
if not os.path.isfile('expressnews.csv'):
    df.to_csv('expressnews.csv', index=False)
else: # else it exists so append without writing the header
    df.to_csv('expressnews.csv', mode='a', header=False, index=False)

#df.to_excel('output.xlsx')

print(df)

    
```
